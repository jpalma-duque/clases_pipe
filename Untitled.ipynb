```python
import matplotlib.pyplot as plt

# Datos para el arroz no modificado genéticamente
tiempo_1 = [0, 2, 4, 6, 8]  # Tiempo en semanas
betacaroteno_1 = [0.4, 0.6, 1.1, 1.3, 1.5]  # Niveles de betacaroteno en microgramos por gramo de arroz

# Datos para el arroz dorado
tiempo_2 = [0, 2, 4, 6, 8]  # Tiempo en semanas
betacaroteno_2 = [0.2, 0.5, 0.8, 1.4, 1.9]  # Niveles de betacaroteno en microgramos por gramo de arroz

# Crear el gráfico de línea
plt.plot(tiempo_1, betacaroteno_1, 'b-', label='Arroz no modificado genéticamente')
plt.plot(tiempo_2, betacaroteno_2, 'orange', label='Arroz dorado')
plt.xlabel('Tiempo (semanas)')
plt.ylabel('Niveles de betacaroteno (u.a.)')
plt.title('Comparación de los niveles de betacaroteno en el arroz dorado y el arroz no modificado genéticamente')
plt.legend()
# Guardar el gráfico como un archivo PNG llamado "comparacion.png"
plt.savefig('comparacion.pdf')
# Mostrar el gráfico
plt.show()


```
